## Regression Data Generation

```python
plt.hist(x,50)
plt.show()
```

```python
y = 3*x + 2
plt.scatter(x,y)
plt.show()
```

## Classification Data Generation

Let us generate some examples of binary classification

```python
u1 = np.array([2,4])
cov_1 = np.array([[2,-.8],
                  [-.8,1]])

u2 = np.array([-1,-3])
cov_2 = np.array([[1,0],
                  [0,2]])
```

```python
X1 = np.random.multivariate_normal(u1,cov_1,300)
print(X1.shape)

X2 = np.random.multivariate_normal(u2,cov_2,400)
print(X2.shape)
```

```python
plt.scatter(X1[:,0],X1[:,1],label="Apple")
plt.scatter(X2[:,0],X2[:,1],marker='^',label="orange")
plt.title("Apples and Oranges")
plt.xlabel("x1 (Size)")
plt.ylabel("x2 (Sweetness)")
plt.legend()
plt.show()
```

```python
y = np.zeros(700)
y[:300] = 1
print(y.shape)
print(y)
```

```python
X = np.vstack((X1,X2))
print(X.shape)
```

```python
#plt.scatter(X[:,0],X[:,1])

for i in range(700):
    if y[i] == 1:
        plt.scatter(X[i,0],X[i,1],color='green')
    else:
        plt.scatter(X[i,0],X[i,1],color='red')
        

plt.show()
```

```python
## Shortcut
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```
